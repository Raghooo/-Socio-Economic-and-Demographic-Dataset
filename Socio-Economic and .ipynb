```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/census_income_data.csv")
```

```python
df.shape
```

```python
df.columns
```

```python
income_map = {'-50000' : 0, ' 50000+.' : 1}
df['income_bracket_numeric'] = df['income_bracket'].map(income_map)
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x='AAGE', y='income_bracket_numeric', data=df)
plt.title('Age Distribution by Income Bracket')
plt.xlabel('AAGE')
plt.ylabel('Income Bracket')
plt.show()
```

```python

plt.figure(figsize=(10, 6))

sns.lineplot(x='AAGE', y='income_bracket_numeric', hue= 'ASEX', data=df)
plt.title('Income Bracket By Sex')
plt.xlabel('AAGE')
plt.ylabel('Income Bracket')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='AHGA', y='income_bracket_numeric', hue = 'ASEX', data=df)
plt.title('Income Bracket By Education')
plt.xlabel('Highest grade attended')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='ACLSWKR', y='income_bracket_numeric', hue = 'ASEX', data=df, palette = 'viridis')
plt.title('Income Bracket By Class_worker')
plt.xlabel('Class_worker')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='AWKSTAT', y='income_bracket_numeric', hue = 'ASEX', data=df, palette= 'YlGnBu')
plt.title('Income Bracket By Work status')
plt.xlabel('Work status')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='AMJIND', y='income_bracket_numeric', hue = 'ASEX', data=df, palette= 'husl')
plt.title('Income Bracket By Major industry')
plt.xlabel('Major industry')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='AMARITL', y='income_bracket_numeric', hue = 'ASEX', data=df, palette= 'viridis')
plt.title('Income Bracket By Marital status')
plt.xlabel('Marital status')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='AMJOCC', y='income_bracket_numeric', hue = 'ASEX', data=df, palette= 'magma')
plt.title('Income Bracket By Major occupation code')
plt.xlabel('Major occupation')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.barplot(x='PRCITSHP', y='income_bracket_numeric', hue = 'ASEX', data=df, palette= 'plasma')
plt.title('Income Bracket By Citizenship status')
plt.xlabel('Citizenship status')
plt.ylabel('Income Bracket')
plt.xticks(rotation=90, ha='right')
plt.show()
```

```python
bins = [0, 1000, 5000, 10000, 50000, 100000, df['CAPGAIN'].max() + 1000]  # Adjusted the last bin edge
labels = ['0-1k', '1k-5k', '5k-10k', '10k-50k', '50k-100k', '100k+']

# Apply binning
df['CAPGAIN_binned'] = pd.cut(df['CAPGAIN'], bins=bins, labels=labels, right=False)

# Check binning
print(df[['CAPGAIN', 'CAPGAIN_binned']])

# Plotting the box plot
plt.figure(figsize=(10, 6))
sns.barplot(x='CAPGAIN_binned', y='income_bracket_numeric', hue='ASEX', data=df, palette='magma')
plt.title('Income Bracket By Capital Gains')
plt.xlabel('Capital Gains')
plt.ylabel('Income Bracket')
plt.xticks(rotation=45, ha='right')
plt.show()
```
